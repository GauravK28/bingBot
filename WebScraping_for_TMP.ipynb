```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```
baseURL = 'https://www1.ncdc.noaa.gov/pub/data/swdi/stormevents/csvfiles/'
page = requests.get(baseURL)
soup = BeautifulSoup(page.text, 'html.parser')
```

```
#list of all csv files
csvFiles = soup('a') #finds all html with <a> tags

#extracting files
StormEventDetails = []
StormEventFatalities = []
StormEventLocations = []

url = 'https://www1.ncdc.noaa.gov/pub/data/swdi/stormevents/csvfiles/StormEvents_details-ftp_v1.0_d1950_c20170120.csv.gz'
#StormEventDetails.append(pd.read_csv(url, compression='gzip'))
#StormEventDetails = pd.concat(StormEventDetails)

for csvFile in csvFiles:
  if csvFile.attrs['href'].find("StormEvents_details") == 0:
    ""
    url = baseURL + csvFile.attrs['href']
    StormEventDetails.append(pd.read_csv(url, low_memory=False))
  if csvFile.attrs['href'].find("StormEvents_fat") == 0:
    StormEventFatalities.append(pd.read_csv(url, low_memory=False))
  if csvFile.attrs['href'].find("StormEvents_loc") == 0:
    StormEventLocations.append(pd.read_csv(url, low_memory=False))

#print(StormEventDetails)
```

```
print("Storm Event Details")
print(StormEventDetails)
print("Storm event Fatalities")
print(StormEventFatalities)
print("Storm event locations")
print(StormEventLocations)
```
